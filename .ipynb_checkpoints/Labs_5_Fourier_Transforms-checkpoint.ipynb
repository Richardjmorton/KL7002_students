# Fourier Transforms

Fourier transforms are one of the most versatile tools, with use and applicability in many areas of science, engineering and beyond. As such, there are many detailed text books that cover the fundamentals and applications. Here, we shall only give a flavour of certain aspects and demonstrate a few examples that are useful for data analysis.

The Fourier transform can be used as an efficient tool to undertake certain computational tasks (e.g. convolution), but also the Fourier transform itself is often of interest (e.g. searching for periodic signals). The Numerical Recipes [1] book gives various examples of its use.

I also recommend the introductory material in [2] and also [3].




## Discrete Fourier transform

Signals are normally discretely sampled and over a finite interval. As such, we can only approximate the signal with a finite number of Fourier terms.

The Discrete Fourier transform, for a series of length $N$ uniformly sampled data points, $h_j$ is


$$
\Large{
H_k = \sum_{j=0}^{N-1} h_j e^{-2\pi i j k\,/N}, 
}
$$
and the inverse is

$$
\Large{
h_j = \frac{1}{N}\sum_{k=0}^{N-1} H_k e^{2\pi i j k\, /N},
}
$$

The discrete Fourier transforms (DFT) are available in a number of packages. We will use `scipy`. In the following we will import the forward and inverse transforms.

````python
from scipy.fftpack import fft, ifft

# create a signal
N = 100
x = np.arange(0, N)
signal = np.sin(2.*np.pi*x / 10)

# Take Fourier transform
ft_signal = fft(signal)

print(ft_signal.shape)

print('DC value: {} '.format(ft_signal[0]))  # f_0 = 0
print('Another value: {} '.format(ft_signal[1]))  # f_1
print('Nyquist value: {} '.format(ft_signal[N//2]))  # f_(n//2+1)
````

We can see a few things after running the above code. The DFT is actual called by FFT, which stands for Fast Fourier Transform - a fast implementation of the DFT ([Cooley and Tukey](https://www.ams.org/journals/mcom/1965-19-090/S0025-5718-1965-0178586-1/S0025-5718-1965-0178586-1.pdf)).

For a time series of length $N$, `scipy.fftpack.fft` returns $N$ Fourier coefficients, $H_k$, in the range $-(N/2-1) < k < (N/2)$. These correspond to the frequencies:

$$
0,\, \frac{1}{NT},\, \frac{2}{NT}, ...,\, \frac{N/2-1}{NT},\, -\frac{1}{2T},\, -\frac{1}{NT}, ...,\, -\frac{N/2-1}{NT}
$$

for even length series, and 

$$
0,\, \frac{1}{NT},\, \frac{2}{NT}, ...,\, \frac{N/2-1}{NT},\, -\frac{1}{NT},\,...,\, -\frac{N/2-1}{NT}  
$$

for odd length series.  $T$ here is the sampling interval.

The Fourier coefficients are also complex numbers, expect for the zero frequency component, called the *'DC component'*, and also the Nyquist Frequency.


For a real valued signal (which is probably always the case in Astrophysics), then $H_{-k}=\bar{H_{k}}$ (the bar means complex conjugate). Hence, the negative frequencies provide no new information.

````python
# example
ft_signal[1:10] == np.flip(np.conj(ft_signal[-9:]))
````

## Examples: signal analysis

### Noise series
Let us look at some examples of using the DFT for analysis of a signal. The search for features (namely periodic ones) in data is a very common task in various areas of astrophysics, e.g. waves in the Sun's atmosphere, X-ray variability of X-ray binaries, detecting gravitational waves. Here we will look at some basic examples of how signals appear in DFTs, in order to build your understanding.

Let us take an example of a time-series called white noise. A white noise signal is when each data point is an uncorrelated random sample with zero mean and finite variance. Uncorrelated means their is no dependence of one sample on another. When the samples are drawn from a Gaussian, it is additive Gaussian white noise.



````python
from scipy.fftpack import fftfreq, fftshift

N = 100  # length of series
dt = 10  # sampling

# generate white noise
series = stats.norm.rvs(random_state=123, size=N)
time = np.arange(0, N, 1)*dt

ft = fft(series) # take DFT
freq = fftfreq(N, d=dt) # Calculate frequencies

# shift both series so negative frequencies come first
freq_s = fftshift(freq)
ft_series = fftshift(ft)

power = np.abs(ft_series)**2 # One-sided PSD

PSD = np.abs(ft_series[0:N//2+1])**2
PSD[1:-1] *= 2  # Two-sided PSD, so multiply by 2, ignoring DC & Nyquist
````

````python
# Plotting
fig, ax = plt.subplots(2, 2, figsize=(10, 7))

ax[0, 0].plot(time, series)
ax[0, 0].set_xlabel('time (s)')
ax[0, 0].set_title("Time-series")


ax[0, 1].plot(freq_s, ft_series.real, label='Real')
ax[0, 1].plot(freq_s, ft_series.imag, label='Imaginary')
ax[0, 1].set_xlabel('Frequency (Hz)')
ax[0, 1].legend()
ax[0, 1].set_title("Fourier components")

ax[1, 0].plot(freq_s, power)
ax[1, 0].set_xlabel('Frequency (Hz)')
ax[1, 0].set_ylabel('Power (units$^2$)')
ax[1, 0].set_title("Power spectra")

ax[1, 1].plot(-freq_s[0:N//2+1], PSD)
ax[1, 1].set_xlabel('Frequency (Hz)')
ax[1, 1].set_ylabel('Power (units$^2$)')
ax[1, 1].set_title("PSD")

plt.tight_layout()
````

What should we expect from a white noise series? 

Given that there should be no particular frequency component that stands out, the power should be distributed evenly across all frequencies.

Looking at the power spectrum plot (lower left), we see the positive and negative frequencies have the same power - as expected for real signals. 

However the power isn't constant! Some frequencies have higher power than the others.

The discrete analogue of Parsvel's theorem (Eq. 5.6) is:

$$
\sum_{j=0}^{N-1} \mid h_j\mid^2 = \frac{1}{N}\sum_{k=0}^{N-1} \mid H_k\mid^2\ = \mbox{Total Power}.
$$

Let left hand side of this can be written as $N\sigma_h^2$ (as the signal has a zero mean), where $\sigma_h$ is the standard deviation of the signal $h$. So we should expect that the power in each frequency bin, for unit variance Gaussian white noise, is $N\sigma_h^2 = N$. 

This suggest that the power spectra is noisy. And for noise processes, it turns out that the power measured at a specific frequency is scattered about its true value following a $\chi^2_2$ distribution (except for the DC and Nyquist frequencies).

We can examine this by comparing to the known distribution.

````python
scaled = PSD[1:-2]/N/2 # divide by two as we multiplied by 2

kernel = stats.gaussian_kde(scaled)
x = np.arange(0, 40, 1)

_ = plt.hist(scaled, density=True, label='Hist', alpha=0.5)
_ = plt.plot(x, kernel(x), label='KDE')

# PDF from chi^2_2
chi2_2 = stats.chi2.pdf(x, df=2)

_ = plt.plot(x, chi2_2, label=r'$\chi_2^2$ PDF')
_ = plt.ylabel('Density')
_ = plt.legend()
````

### Harmonic series

In the following example, we will see what happens when we take the DFT of a sinusoid. We will look at two illustrative examples. But first we note that the Fourier transform of a sinusoid is the sum of two delta functions, i.e.

$$
\mathcal{F}(sin(2\pi f_0 t)) = \frac{1}{2}\left[\delta(f-f_0)+\delta(f+f_0) \right].
$$

The sinusoid and the delta functions are known as a Fourier transform pair. For some common functions, there are [known Fourier transform pairs](https://en.wikipedia.org/wiki/Fourier_transform#Tables_of_important_Fourier_transforms).

**Example 1**
The first case of interest is comparing two signals. The signals are of the same length and sampled the same, but one has an integer number of periods in the period of measurement and the other has a non integer number of periods.

````python
dt = 1./128 # sampling

time = np.arange(0, 1, dt)
N = len(time)
f = [3., 3.5] # frequency

fig, ax = plt.subplots(2,2, figsize=(10,7))

for fr in f:
    
    series = np.sin(2.*np.pi*fr* time)
    ax[0,0].plot(time, series, label=str(fr))
    ax[0,0].set_xlabel('Time (s)')
    
    ft = fft(series) # take DFT
    ft_series = fftshift(ft) # calculate frequecnices

    freq = fftfreq(N, d = dt)
    freq_s = fftshift(freq)
    
    PSD = np.abs(ft_series[0:N//2+1])**2 # power
    PSD[1:-1] *= 2.# two-sided
    
    ax[0,1].plot(-freq_s[0:N//2+1], PSD,'o', label=str(fr))
    ax[0,1].set_xlabel('Frequency (Hz)')
    ax[0,1].set_ylabel(r'units$^2$')
    ax[0,1].set_xlim(0,20)
    
    ax[1,0].plot(-freq_s[0:N//2+1], PSD, label=str(fr))
    ax[1,0].set_xlabel('Frequency (Hz)')
    ax[1,0].set_ylabel(r'units$^2$')
    ax[1,0].set_yscale('log')
    


ax[0,0].legend()
ax[0,1].legend()
ax[1,0].legend()
plt.tight_layout()
````

We see that the signal with an integer number of periods (blue) is essentially a $\delta$ function (only one of the two $\delta$ functions is shown here as the figure is the two sided power spectrum), with all the power concentrated at the corresponding frequency location. The signal with non-integer number of periods has its power spread out over all the neighbouring frequencies.

Why does this occur? The DFT essentially treats the data as if it were periodic outside of the measurement window. And the DFT is evaluated at the fundamental frequency of the measurement window, $\frac{1}{N\Delta}$ Hz and its harmonics.

For non-integer cycles, the power is smeared over all frequencies, with effect being called 'spectral leakage'. The leakage of power is much clearer in the log axis plot. The series with a non-integer number of cycles has a discontinuity at the edge of the measurement window, and the DFT places power at other frequencies in an effort to describe the discontinuity.

We are also stuck with a problem that we can't estimate the true frequency of the signal properly, as its frequency lies between two of the frequency bins.

There are methods available to us that can be used to help overcome both of the problems, called *windowing* and *zero-padding*. We will discuss them in the exercises. 

We can also use the DFT results find the amplitude of the signal. The two-sided amplitude spectrum (with the units, $units_{rms}$) is given as:
$$
A = \sqrt{2}\frac{\mid H_k \mid}{N},
$$
which gives the root mean squared amplitude for each sinusoidal component. Note DC and Nyquist components are not multiplied by the $\sqrt{2}$.

**Example 2**

In the second example, we will look at the same signal sampled at two different rates.

````python
N = [1000, 100] # number of samples
dt = [1, 10] # sampling
markers = ['+','*']

for Ns, dts, mrk in zip(N,dt, markers):
    time = np.arange(0,Ns,1)*dts
    series = np.sin(2.*np.pi*time/100)
    
    ft = fft(series)
    ft_series = fftshift(ft)

    freq = fftfreq(Ns, d = dts)
    freq_s = fftshift(freq)

    PSD = np.abs(ft_series[0:Ns//2+1])**2
    PSD[1:-1] *= 2
    plt.plot(-freq_s[0:Ns//2+1], PSD, mrk, label='sampling {:1.0f}'.format(dts))
    plt.xscale('log')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel(r'units$^2$')
    
    print('sampling = {0:1.2f}, total power ={1:1.2f}'.format(dts, PSD.sum()/Ns))
    print('Series {:1.2f}'.format((series**2).sum()))
    
    
_ = plt.legend()
````

We have plotted the power for both series, however, things are not quite as you might expect. Even though both signals have the same amplitude, and should have the same power, the signal with faster sampling has a much greater power. This situation does not permit us to compare the same phenomenon which happened to be sampled at different rates.

In order to be able to compare signals, we have to normalise the power spectra to take into account the difference in sampling. The correct normalisation is

$$
PSD(f_k) = \frac{2 \Delta}{N}\mid H_k \mid^2 
$$

(The factor of 2 does not apply for the DC or Nyquist components).


````python
N = [1000, 100] # number of samples
dt = [1, 10] # sampling
markers = ['+','*']

for Ns, dts, mrk in zip(N,dt, markers):
    time = np.arange(0,Ns,1)*dts
    series = np.sin(2.*np.pi*time/100)
    
    ft = fft(series)
    ft_series = fftshift(ft)

    freq = fftfreq(Ns, d = dts)
    freq_s = fftshift(freq)

    PSD = np.abs(ft_series[0:Ns//2+1])**2 *(dts / Ns)
    PSD[1:-1] *= 2
    plt.plot(-freq_s[0:Ns//2+1], PSD, mrk, label='sampling {:1.0f}'.format(dts))
    plt.xscale('log')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel(r'units$^2 / Hz$')
    
    print('sampling = {0:1.2f}, total power = {1:1.2f}'.format(dts, PSD.sum()))
    print('Series {:1.2f}'.format((series**2).sum()))
    
    
_ = plt.legend()
````

## Examples: Cross-correlation

Another useful application for Fourier transforms is the registration of images. Typically in astrophysics, and especially in solar physics, the analysis of data requires that you have images that are perfectly aligned. 

Let us picture the following example. A satellite, such as the Solar Dynamics Observatory, is taking images of the Sun. The satellite is moving around Earth, and is constantly trying to keep the Sun in the main focus of the telescope. This does not happen perfectly and so each subsequent image will be displaced in some manner. Hence, a feature of interest on the Sun will move around the camera sensor. 

While this may not matter for all science objectives, some tasks, such as measuring the physical movement or velocity of feature, requires you to know the position of the feature precisely.

So, how do we find the offsets between two images?

One method is to use cross-correlation. If we look at the formula for cross-correlation

$$
h\star g = \int_{-\infty}^{\infty}h(\tau+t)g(\tau)\,d\tau.
$$ 

we see that for two signals, this formula measures essentially how alike the signals are for different values of the lag, $t$. The lag here is essentially the off-set between two signals.

The correlation function behaves essentially like the figures, it takes $h$ and slides it along $g$ at fixed lag values ($t$), calculating the correlation between the signals for each lag. So to find the offset between two signals, you calculate the correlation function and find the value of lag which gives the largest correlation. This then tells you the offset between two series.

Let's try this for a pair of images.

````python
# import and plot image
from astropy.io import fits
import sunpy.visualization.colormaps
from os import path

cmap = plt.get_cmap('sdoaia171') 

# download file from KL7002 GitHub site
#!wget https://raw.githubusercontent.com/Richardjmorton/KL7002_students/main/datasets/ar_cut_out.fits


filename=path.expanduser('~')+'/github_reps/KL7002/datasets/ar_cut_out.fits'
# For Colab use:
#filename='/content/ar_cut_out.fits'

with fits.open(filename) as file:
    data = file[0].data

plt.figure(figsize=(10,10))
_ = plt.imshow(data**0.2, origin='lower', cmap = cmap)
````

First let us shift the figures functions in scipy.

````python
from scipy.fftpack import fft2, ifft2
````

````python
# create shifted images

from scipy.ndimage import fourier_shift

shift = (-33.3, 15.71)
# The shift corresponds to the pixel offset relative to the reference image

offset_image = fourier_shift(fft2(data), shift)
offset_image = abs((ifft2(offset_image)))

fix, ax = plt.subplots(1, 2, figsize=(15,10))
_ = ax[0].imshow(data**0.2, origin='lower', cmap = cmap)
_ = ax[1].imshow(offset_image**0.2, origin='lower', cmap = cmap)
````

Then, we can calculate the shift using the Correlation Theorem, i.e. take the Fourier transform, multiply the data and invert to find the 2D correlation function. The maximum location of the correlation function should give the shifts to the nearest pixel value.

````python
# Fourier transform of images
fft_orig = fft2(data)
fft_shift = fft2(offset_image)

# calculate correlation function
corr = fft_orig*np.conj(fft_shift)
corr_shift = fftshift(ifft2(corr))

_ = plt.imshow(corr_shift.real)
_ = plt.title("2D correlation function")
_ = plt.colorbar()

#look for max of correlation function
vals = np.unravel_index(np.argmax(corr_shift.real), corr_shift.shape)

lenx, leny = corr_shift.shape
print('Shift is ({0:1.1f}, {1:1.1f})'.format(vals[0]-lenx//2, vals[1]-leny//2))
````

Not a bad estimate! There are more [advanced methods](https://scikit-image.org/docs/stable/auto_examples/registration/plot_register_translation.html) that can be used to calculate the shift to a higher degree of accuracy. We will look at a basic version in the exercises.

3. Here we will look at an aspect called 'zero-padding' of a signal, and see the effect it has on the power spectra. Zero padding is essentially adding some number of zeros to the end of your signal.

   After zero-padding, a signal given to the FFT is now longer. 4. One aspect we haven't discussed yet is the window of the observations. The DFT makes the assumption that the signal we observe continues outside of our observation window for infinite time and is periodic at intervals over the length of the window. So, we can think that our observations are defined by a rectangular window that is of length N (see plot below). Hence, in the time domain the measured signal is a product of the true signal, $h$ and the window, $w$. In the Fourier domain it is the convolution of these two signals, i.e.
   $$
   hw \Longleftrightarrow H*W.
   $$

   For a rectangular window, the Fourier transform is the $\mathrm{sinc}$ function, i.e.

   $$
   \mathcal{F}( \mathrm{rect}(at) ) = \frac{1}{\mid{a}\mid}\mathrm{sinc\left(\frac{f}{a}\right)}=\frac{1}{\mid{a}\mid}\frac{sin\left(\pi\frac{f}{a}\right)}{\pi\frac{f}{a}}.
   $$

   Hence, for the sinusoid example in Question 1, we should expect that we have a pair of $\delta$ functions convolved with the $\mathrm{sinc}$ function; which is what we observe. The impact of the window on the signal is also a form of 'leakage'. In order to minimise the leakage, various [window functions](https://en.wikipedia.org/wiki/Window_function#Windowing) have been proposed, each with there own pro's and con's. Note, the leakage is zero when the function has an integer number of periods within the window. I suggest checking out the excellent graphic [here](https://jackschaedler.github.io/circles-sines-signals/dft_leakage.html) which demonstrates why (Figure 3).

````python
# window plot
N = 200
x = np.arange(0, N, 1)
signal = 0.5 + 0.25*np.sin(2*np.pi*x/10)
window = np.zeros(N)
window[N//4:3*N//4] = 1

_ = plt.plot(signal, label='signal $h$')
_ = plt.plot(window,  label=r'window $w$')
_ = plt.fill(window, color='gray', alpha=0.5,
       label='Windowed part of signal')
_ = plt.ylim(-0.5,1.7)
_ = plt.xlabel('time')
_ = plt.legend(loc='upper right')
````
